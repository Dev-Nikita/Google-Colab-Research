```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# For fuzzy AHP
from math import prod

criteria = ["Quality", "Speed", "Cost", "Flexibility"]

A = np.array([
    [1,   5,   3,   3],
    [1/5, 1,   1/3, 1/2],
    [1/3, 3,   1,   2],
    [1/3, 2,   1/2, 1]
])

A

col_sum = A.sum(axis=0)
norm_matrix = A / col_sum
weights = norm_matrix.mean(axis=1)

df_weights = pd.DataFrame({"Criterion": criteria, "Weight": weights})
df_weights

Aw = A @ weights
lambda_max = (Aw / weights).mean()
CI = (lambda_max - len(criteria)) / (len(criteria) - 1)
RI = 0.90  # for n=4
CR = CI / RI

lambda_max, CI, CR

tech = [
    "3D FDM", "3D SLA", "Electronic",
    "Manual", "Mechanical", "Thermoforming", "Traditional"
]

fuzzy_data = np.array([
    [0.5, 0.5, 0.7, 0.8],
    [0.8, 0.4, 0.3, 0.8],
    [0.8, 0.9, 0.1, 0.8],
    [0.5, 0.1, 0.9, 0.5],
    [0.8, 0.5, 0.5, 0.8],
    [0.8, 0.3, 0.5, 0.3],
    [0.8, 0.8, 0.6, 0.8],
])

final_scores = fuzzy_data @ weights
df_scores = pd.DataFrame({"Technology": tech, "Score": final_scores})
df_scores.sort_values("Score", ascending=False)

plt.figure(figsize=(10,6))
plt.barh(df_scores["Technology"], df_scores["Score"])
plt.xlabel("Aggregated Priority")
plt.title("Final Ranking of Braille Technologies (AHPFuzzy)")
plt.gca().invert_yaxis()
plt.show()

plt.figure(figsize=(8,6))
plt.imshow(fuzzy_data, cmap="Blues")
plt.xticks(range(len(criteria)), criteria)
plt.yticks(range(len(tech)), tech)
plt.colorbar(label="Fuzzy Value")
plt.title("Fuzzy Evaluation Matrix")
plt.show()

w_range = np.linspace(0.3, 0.8, 20)
sens_results = []

criteria_names = ["Quality", "Speed", "Cost", "Flexibility"]

#   
w_values = np.linspace(0.1, 0.7, 30)

#   ( AHP)
base = np.array([0.50, 0.08, 0.26, 0.16])

sens_total = {}

for idx, cname in enumerate(criteria_names):
    sens_results = []
    for w in w_values:
        wc = base.copy()
        wc[idx] = w
        wc = wc / wc.sum()   # 
        score = fuzzy_data @ wc
        sens_results.append(score)
    sens_total[cname] = np.array(sens_results)


for cname in criteria_names:
    plt.figure(figsize=(10,6))
    for i, tech_name in enumerate(tech):
        plt.plot(w_values, sens_total[cname][:, i], label=tech_name)

    plt.title(f"Sensitivity Analysis  Influence of '{cname}' Weight")
    plt.xlabel("Weight of Criterion")
    plt.ylabel("Final Score")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

```
