```python
# =============================
# Braille Technologies Clustering Figure (Springer-ready PNG)
# Google Colab code
# =============================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# --- Data from your study (defuzzified values + weights)
# Technologies:
# Traditional, Mechanical, SLA, Electronic, Thermoforming, FDM, Manual

technologies = [
    "Traditional embossing",
    "Mechanical embossing",
    "3D printing (SLA/DLP)",
    "Electronic displays",
    "Thermoforming",
    "3D printing (FDM)",
    "Manual methods"
]

# Defuzzified aggregated scores by criterion (D values)
# Order: Quality, Speed, Cost, Flexibility
data = np.array([
    [0.8, 0.8, 0.6, 0.8],   # Traditional
    [0.8, 0.5, 0.5, 0.8],   # Mechanical
    [0.8, 0.4, 0.3, 0.8],   # SLA
    [0.8, 0.9, 0.1, 0.8],   # Electronic
    [0.8, 0.3, 0.5, 0.3],   # Thermoforming
    [0.5, 0.5, 0.7, 0.8],   # FDM
    [0.5, 0.1, 0.9, 0.5],   # Manual
])

# Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# PCA reduction to 2D
pca = PCA(n_components=2)
reduced = pca.fit_transform(scaled_data)

# Plot
plt.figure(figsize=(8, 6), dpi=300)
plt.scatter(reduced[:,0], reduced[:,1], s=120, color='black')

for i, tech in enumerate(technologies):
    plt.text(reduced[i,0] + 0.03, reduced[i,1] + 0.03, tech, fontsize=8)

plt.title("Clustered Distribution of Braille Formation Technologies\n(PCA-based reduction, 4 weighted criteria)", fontsize=10)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True, linewidth=0.3)
plt.tight_layout()

# Save as PNG
plt.savefig("braille_technology_clusters.png", dpi=300)

plt.show()
```
